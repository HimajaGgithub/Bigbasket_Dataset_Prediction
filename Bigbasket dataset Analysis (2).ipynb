## Labelling the dataset

```python
def discount_product(data):
    return ((data['market_price'] - data['sale_price']))
def is_discount_product(data):
    return ((data['market_price'] - data['sale_price']) != 0)


df = df.assign(discount_product=df[['sale_price', 'market_price']].apply(discount_product,axis=1))
df = df.assign(is_discount_product=df[['sale_price', 'market_price']].apply(is_discount_product,axis=1))
df
```
