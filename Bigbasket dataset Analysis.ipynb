## Analysis of Duplicates

```python
#To find the duplicates, the index column has to be made Zero
data_bb2=pd.read_csv('C:/Users/HIMAJA/OneDrive/Desktop/drive/dataset_bigbasket1.csv',index_col=0)
```

```python
#Duplicates details
duplicate = data_bb2.duplicated()
duplicate_count = sum(data_bb2.duplicated())

print("There are", duplicate_count, "duplicates in the data set.\n")

# Shows all duplicate data
if duplicate_count > 0:
    display("DUPLICATE DATA:\n", data_bb2[duplicate])
```

```python
#Number of duplicates
data_bb2.duplicated().sum()
```

```python
#DUPLICATES REMOVED
data_bb2.drop_duplicates(inplace=True)
data_bb2.duplicated().sum()
```

## Dealing with Null values

```python
#Null value is represented in percentage
percent_missing = data_bb2.isnull().sum() * 100 / len(data_bb2)
missing_value_df = pd.DataFrame({'column_name': data_bb2.columns,
                                 'percent_missing': percent_missing})
missing_value_df
```

```python
print(data_bb2.isnull().sum())
```

```python
#Removing null values by filling it with the mean of all values
data_bb2['rating'].fillna(data_bb2.groupby("category")['rating'].transform("mean"), inplace=True)
```

```python
percent_missing = data_bb2.isnull().sum() * 100 / len(data_bb2)
missing_value_df = pd.DataFrame({'column_name': data_bb2.columns,
                                 'percent_missing': percent_missing})
missing_value_df
```

```python
#Check for remaining null values
print(data_bb2.isnull().sum())
```

```python
#Dropping the remaining null values
df=data_bb2.dropna()
```

```python
print(df.isnull().sum())
```

```python
display(df)
```

## Labelling the dataset

```python
def discount_product(data):
    return ((data['market_price'] - data['sale_price']))
def is_discount_product(data):
    return ((data['market_price'] - data['sale_price']) != 0)


df = df.assign(discount_product=df[['sale_price', 'market_price']].apply(discount_product,axis=1))
df = df.assign(is_discount_product=df[['sale_price', 'market_price']].apply(is_discount_product,axis=1))
df
```

## Naming convenctions

```python
#Checking for naming convenctions
for column in df.columns:
    if not column.isalnum():
        print('Found typo error in column:', column)

for column in df.columns:
    if not column.startswith('lower_case'):
        print('Found naming convention error in column:', column)
```

```python
#Capitalizing the column names
columns = df.columns
capitalized_columns = [c.capitalize() for c in columns]
df.columns = capitalized_columns
display(df)
```

```python
df['Category']=df['Category'].str.replace('&','and')
df['Sub_category']=df['Sub_category'].str.replace('&','and')
df['Type']=df['Type'].str.replace('&','and')
display(df)
```

## Validation of dataset

```python
sales_price = df["Sale_price"]
market_price = df["Market_price"]

# Check if sales price is lesser than market price
is_lesser = sales_price <= market_price

if(is_lesser.sum()==len(df)):
    print("The given dataset is VALIDATED because the Sales price of all the products is lesser than or equal to the market price of that product!")
else:
    print("The dataset is not validated. Please VALIDATE")
```
